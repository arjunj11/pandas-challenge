```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python

purchase_data2= purchase_data.drop_duplicates(subset="SN")
bins=[0,10,15,20,25,30,35,40,120]
group_labels=[]
for i in range(len(bins)-1):
    group_labels.append(f"{bins[i]} to {bins[i+1]-1}")
group_labels[len(bins)-2]=">40"
group_labels[0]="<10"
group_labels

purchase_data2["Age Demo"]=pd.cut(purchase_data2["Age"],bins,labels=group_labels,right=False)

summary_data=purchase_data2.groupby("Age Demo").count()
summ_data=summary_data.rename(columns={"Purchase ID":"Total Counts","SN":"Percentage of Players"})
summ_data["Percentage of Players"]=summ_data["Total Counts"]*100/summ_data["Total Counts"].sum()

summary_final=summ_data[["Total Counts","Percentage of Players"]]
summary_final["Percentage of Players"]=summary_final["Percentage of Players"].astype(float).map("{:,.2f}%".format)

summary_final
                        

```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
sum1_data=purchase_data.groupby("SN").sum()
count1_data=purchase_data.groupby("SN").count()
final1_data=count1_data
final1_data["Purchase Count"]=count1_data["Purchase ID"]
final1_data=final1_data.drop(columns=["Purchase ID","Age","Gender","Item ID","Item Name","Price","Age Demo"])
final1_data["Total Purchase Value"]=sum1_data["Price"]
final1_data["Average Purchase Price"]=final1_data["Total Purchase Value"]/final1_data["Purchase Count"]
final_data=final1_data.sort_values(['Total Purchase Value'], ascending=False)
final_data["Total Purchase Value"]=final_data["Total Purchase Value"].astype(float).map("${:,.2f}".format)
final_data["Average Purchase Price"]=final_data["Average Purchase Price"].astype(float).map("${:,.2f}".format)
final_data.head()
```
